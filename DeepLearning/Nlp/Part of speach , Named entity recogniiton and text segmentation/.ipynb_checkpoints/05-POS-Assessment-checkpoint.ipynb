**6. Show the first two named entities from Beatrix Potter's *The Tale of Peter Rabbit* **

```python
def show_ents(doc):
    if doc.ents:
        for ent in doc.ents[:2]:
            print(ent.text+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
    else:
        print('No named entities found.')
show_ents(doc)
```
